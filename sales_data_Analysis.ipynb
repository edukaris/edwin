```python
df['Month']= df['Order Date'].str[0:2]
```

### Clean up the data

#### Drop all Nan rows

```python
df.dropna(inplace=True)
```

#### check data with 'or' values

```python
df = df[df['Month'].str[0:2]!='Or']
df
```

### Add a sale_value column

```python
df['Quantity Ordered']= df['Quantity Ordered'].astype('int32')
df['Price Each']=df['Price Each'].astype('float')
df['sale_value']= df['Quantity Ordered']*df['Price Each']
df
```

### Add city Column

```python
df
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2][1:3]

df['City'] = df['Purchase Address'].apply(lambda x: get_city(x) + '(' + get_state(x) + ')')
df
```

```python
highest_city = df.groupby(['City']).sum()
highest_city
```

```python
fig, ax = plt.subplots(figsize=(15,12))
plt.bar(highest_city.index.values, highest_city['sale_value'])
# plt.yticks(highest_month['sale_value'])
plt.xticks(highest_month.index.values, rotation=45)
plt.ylabel("sales in Millions($) ")
plt.xlabel('City')
```

### What time should you display adverts to increase effectiveness and increase sales

```python
df.head()
```

```python
df['Order Date']= pd.to_datetime(df['Order Date'])
df
```

```python
df['hour']= df['Order Date'].dt.hour
df['minute']= df['Order Date'].dt.minute
df.head()
```

```python
highest_hour = df.groupby(['hour']).count()
highest_hour
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(highest_hour.index.values, highest_hour['sale_value'],color='red', marker='*')
plt.xticks(highest_hour.index.values, rotation=90)
plt.ylabel("number of orders",fontsize=14)
plt.xlabel('Hour of the day.',fontsize=14)
plt.title('Order numbers Vs Hour of the day', fontsize=14,  fontweight='bold')
plt.grid(True)
```

```python
# I recommend targeting the ads around between 1100hrs and 1200hrs and at around 1800hrs and 1900hrs
```

### What products are most often sold together?

```python
df.head()
```

```python
new_df = df[df['Order ID'].duplicated(keep=False)]
new_df
```

```python
new_df['grouped']= new_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
new_df = new_df[['Order ID', 'grouped']].drop_duplicates()
new_df
```

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in new_df['grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
count.most_common(20)
```
